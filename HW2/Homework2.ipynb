### Part 3 - Basic data analysis with Python

Here is a dataset of Seattle weather in 4 years. Load the data and generate corresponding data visualizations. Make sure each figure has title, labels and legends if necessary. Add one or two sentences in each figure to talk about what you have learned from the figure. 

```python
from vega_datasets import data

weather = data.seattle_weather()
weather.head()
```

##### Question 1 (5 pts)

Generate a data visualization to show the relationship between precipitation and max temperature.  

```python
import matplotlib.pyplot as plt 

plt.scatter(weather['precipitation'], weather['temp_max'])

plt.xlabel('Precipitation')
plt.ylabel('Max Temperature')
plt.title('Scatter plot of precipitation and max temperature in Seattle')

plt.show()
```

The plot shows that there is no clear relationship between precipitation and max temperature in Seattle. Max temperature goes from below 0 to slightly above 35 degrees with no outliers, while precipitation has a few outliers that are above 50.

##### Question 2 (5 pts)

Generate a data visualization to show the relationship between max temperature and weather type.  

```python
# group by weather type and get the means of the max temperatures for each weather type
df_weather_type = weather.groupby('weather').mean().reset_index()

plt.bar(x = df_weather_type['weather'], height = df_weather_type['temp_max'])

plt.xlabel('Weather Type')
plt.ylabel('Average Max Temperature')
plt.title('Bar plot of weather type and average max temperature in Seattle')

plt.show()
```

The figure shows that the sun weather type has the highest average maxmium temperature in Seattle, being just under 20 degrees on average. Snow has by far the lowest average maximum temperature, while drizzle, fog and rain all have similar average temparatures. 

##### Question 3 (5 pts)

Generate a data visualization to show the relationship between max temperature and min temperature, and then show different color for differnet weather type. 

```python
import seaborn as sns
sns.scatterplot(data = weather, x = 'temp_max', y = 'temp_min', hue='weather')

plt.xlabel('Max Temperature')
plt.ylabel('Min Temperature')
plt.title('Scatter plot of max temperature and min temperature according to weather type in Seattle')

plt.legend(title='Weather Type')

plt.show()
```

The scatterplot shows that the sun and rain weather types have a large range of max and min temperatures, while snow only has low tempatures. It also shows that Max Temperature and Min Temperature have a positive linear relationship, which makes sense.

##### Question 4 (5 pts) 

Generate a data visualization to show the relationship between max temperature, min temperature and wind speed. 

```python
sns.scatterplot(data = weather, x = 'temp_max', y = 'temp_min', hue='wind')


plt.xlabel('Max Temperature')
plt.ylabel('Min Temperature')
plt.title('Scatter plot of max temperature and min temperature based on wind speed in Seattle')

plt.legend(title='Wind Speed')

plt.show()
```

The graph shows that high wind speeds tend to be associated with higher minimum temperatures and lower maxiumum temperatures, meaning that when wind speed is higher, there is less variation between min and max temperature. Also, at the higher temperatures, wind speed seems to hover around 2 or 4 and rarely gets up to over 8.

### Part 4 - Basic data analysis with Altair

We are going to use the same data but use Altair to create the data visualziation. 

##### Question 1 (5 points)

Recreate the figure in Part 3 Question 2 with Altair. 

```python
import altair as alt

alt.Chart(weather).mark_bar().encode(
    alt.Y('average(temp_max):Q', title='Average Max Temp'),
    alt.X('weather:N', title='Weather Type')
).properties(width=325, height=300, title='Bar Plot of Average Max Temp by Weather Type in Seattle')
```

##### Question 2 (5 points)

Recreate the figure in Part 3 Question 3 with Altair.

```python
alt.Chart(weather).mark_circle().encode(
    alt.Y('temp_min:Q', title='Min Temp'),
    alt.X('temp_max:Q', title='Max Temp'),
    alt.Color('weather:N')
).properties(width=325, height=300, 
             title='Scatter plot of max temperature and min temperature based on weather type in Seattle')
```

##### Question 3 (10 points)

Create a facet figure with 2 row and 3 columns. For each panel in the facet figure, it shows a scatter plot between the max and min temperature for one type of weather. Use the wind speed to color the points. Also add interaction to show the max and min temperature and wind speed for each data point. Include necessary title and labels. 

```python

alt.Chart(weather).mark_circle().encode(
    alt.Y('temp_min:Q', title='Min Temp'),
    alt.X('temp_max:Q', title='Max Temp'),
    alt.Tooltip(['temp_max:Q','temp_min:Q', 'wind:Q'])
).properties(width=325, height=300).facet(
                 facet = 'weather:N', columns=3, title='Relationship between weather type and max and min temperature in Seattle')
```

##### Question 4 (10 points)

Create a compound figure with at least two types of marks. Show the trend of average temperature (the average of min and max temperature) in 4 years. Include necessary title and labels.

```python
weather['temp_avg'] = (weather['temp_max'] + weather['temp_min'])/2
weather.head()
```

```python
chart_max = alt.Chart(weather).mark_circle(
    color = 'red'
).encode(
    alt.Y('temp_max:Q', title='Max Temp'),
     alt.X('date:T', title='Date')
)

chart_min = alt.Chart(weather).mark_circle(
    color = 'blue'
).encode(
    alt.Y('temp_min:Q', title='Min Temp'),
    alt.X('date:T', title='Date'),
)

chart_avg = alt.Chart(weather).mark_line(
    color = 'black',
    strokeWidth=3,
    interpolate = 'monotone'
).encode(
    alt.Y('temp_avg:Q', title='Avg Temp'),
     alt.X('date:T', title='Date')
)

chart = chart_max+chart_min+chart_avg
chart.properties(width=300, height=300, title='Maximum, Minimum, and Average Temperature for the last 4 years in Seattle')
```
